```python
import pandas as pd

# Step 1: Define the selected best categories
selected_categories = [
    'Electronics_5',
    'Books_5',
    'Clothing_Shoes_and_Jewelry_5',
    'Home_and_Kitchen_5',
    'Kindle_Store_5'
]

# Step 2: Filter DataFrame to only include selected categories
filtered_df = df[df['category'].isin(selected_categories)]

# Step 3: For each category, sample 1000 rows
sampled_dfs = []
for category in selected_categories:
    category_df = filtered_df[filtered_df['category'] == category]

    # In case a category has fewer than 1000 rows, sample with replacement
    sample_df = category_df.sample(n=1000, replace=(len(category_df) < 1000), random_state=42)
    sampled_dfs.append(sample_df)

# Step 4: Concatenate the sampled DataFrames
final_df = pd.concat(sampled_dfs).reset_index(drop=True)

# Step 5: Shuffle the final dataset
final_df = final_df.sample(frac=1.0, random_state=42).reset_index(drop=True)

# Optional: Save to CSV for future use
final_df.to_csv("filtered_balanced_reviews.csv", index=False)

# Check
print(final_df['category'].value_counts())
print(final_df.shape)
```

```python
import pandas as pd
import spacy
from sentence_transformers import SentenceTransformer
import torch

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Check if GPU is available for embeddings
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("Device being used:", device)
```

```python
def preprocess(text):
    doc = nlp(text.lower())
    return ' '.join([token.lemma_ for token in doc if token.is_alpha and not token.is_stop])

final_df['clean_text'] = final_df['text_'].apply(preprocess)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('stsb-roberta-base', device=device)

# Batch encode with GPU
embeddings = model.encode(
    final_df['clean_text'].tolist(),
    batch_size=64,
    show_progress_bar=True,
    device=device
)

import numpy as np
X = np.array(embeddings)
y = final_df['label'].map({'CG': 1, 'OR': 0}).values  # Binary label: 1 = fake, 0 = real

```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from tqdm import tqdm

# Fitness function: lower score = better (error = 1 - accuracy)
def fitness(mask, X, y):
    if np.count_nonzero(mask) == 0:
        return 1.0  # all-zero mask = invalid
    selected = X[:, mask == 1]
    clf = RandomForestClassifier(n_estimators=50)
    score = cross_val_score(clf, selected, y, cv=3, scoring='accuracy')
    return 1 - np.mean(score)

# TSA update logic
def TSA_optimize(X, y, num_agents=10, max_iter=15):
    dim = X.shape[1]
    # Initialize population with binary vectors
    population = np.random.randint(0, 2, (num_agents, dim))
    fitness_vals = np.array([fitness(ind, X, y) for ind in population])
    best_idx = np.argmin(fitness_vals)
    best_agent = population[best_idx].copy()

    for it in tqdm(range(max_iter), desc="TSA Optimization"):
        for i in range(num_agents):
            r1 = np.random.rand()
            r2 = np.random.rand()
            new_agent = population[i].copy()
            for d in range(dim):
                if np.random.rand() < 0.5:
                    new_agent[d] = best_agent[d]  # attract toward best
                else:
                    new_agent[d] = np.random.randint(0, 2)  # random mutation
            # Evaluate
            new_fit = fitness(new_agent, X, y)
            if new_fit < fitness_vals[i]:
                population[i] = new_agent
                fitness_vals[i] = new_fit
                if new_fit < fitness(best_agent, X, y):
                    best_agent = new_agent
        # Optional: print intermediate best accuracy
        print(f"Iter {it+1} | Best Accuracy: {1 - fitness(best_agent, X, y):.4f}")

    return best_agent
```

```python
# Load if needed
# X = np.load("X_roberta_embeddings.npy")
# y = np.load("y_labels.npy")

# Run TSA
best_mask = TSA_optimize(X, y, num_agents=10, max_iter=15)
```

```python
num_selected_features = np.sum(best_mask)
print(f"Number of selected features: {num_selected_features}")
```
